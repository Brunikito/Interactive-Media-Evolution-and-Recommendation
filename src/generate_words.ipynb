```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://greenopolis.com/adjectives-list/'
response = requests.get(url)
if response.status_code == 200:
    page_content = response.content
else:
    print(f"Erro ao acessar a página: {response.status_code}")
```

```python
soup = BeautifulSoup(page_content, 'html.parser')
```

```python
# Foca só na parte principal do artigo
main_content = soup.find('div', class_='entry-content')

adjetivos = []
if main_content:
    for ul in main_content.find_all('ul'):
        for li in ul.find_all('li'):
            texto = li.get_text(strip=True)
            # Ignora títulos e links que não são adjetivos
            if texto and len(texto.split()) == 1 and texto.isalpha():
                adjetivos.append(texto.lower())  # converte para minúsculo, se quiser
```

```python
adjetivos_unicos = sorted(set(adjetivos))
```

```python
print(adjetivos_unicos)
```

```python
with open('../data/words/adjectives.txt', 'w') as f:
    for adjetivo in adjetivos_unicos:
        f.write(f"{adjetivo}\n")
```

```python
url = 'https://greenopolis.com/list-of-nouns/'
response = requests.get(url)
```

```python
if response.status_code == 200:
    page_content = response.content
else:
    print(f"Erro ao acessar a página: {response.status_code}")
```

```python
soup = BeautifulSoup(page_content, 'html.parser')
```

```python
main_content = soup.find('div', class_='entry-content')

substantivos = []
if main_content:
    for ul in main_content.find_all(['ul', 'ol']):
        for li in ul.find_all('li'):
            substantivo = li.get_text(strip=True)
            # Verificar se o texto é uma única palavra e contém apenas letras
            if substantivo.isalpha():
                substantivos.append(substantivo.lower())  # Opcional: converter para minúsculas
```

```python
substantivos_unicos = sorted(set(substantivos))
```

```python
print(substantivos_unicos)
```

```python
with open('../data/words/nouns.txt', 'w') as f:
    for substantivo in substantivos_unicos:
        f.write(f"{substantivo}\n")
```
